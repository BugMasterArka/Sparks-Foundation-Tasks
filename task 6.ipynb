# Decision Tree Classification

## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import tree
```

## Importing the Dataset

```python
df = pd.read_csv('D:/Sparks Foundation Tasks/Iris.csv')
df.head()
```

## Dividing the Dataset into Independent and Dependent Variable

```python
X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

## Label Encoding the Dependent Variable

```python
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

## Dividing the dataset into Training and Test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Training the model

```python
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

## Evaluating our model

```python
y_pred = classifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Predicting a new Result

```python
print(classifier.predict([[5.8,3.3,4.7,1.4]]))
```

## Printing the Decision Tree (Method 1)

```python
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)
tree.plot_tree(classifier, feature_names = fn, class_names=cn,filled = True);
```
